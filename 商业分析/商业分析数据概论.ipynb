```python
import pandas_datareader as web 
import datetime

#定义获取数据的时间段
#开始时间为2022，1，1
start=datetime.datetime(2022,1,1)
end=datetime.date.today() #获取结束时间

#采用查询stooq股指数据dji道琼斯工业股
stock=web.DataReader('^DJI','stooq',start,end)
#对数据进行逆转，计算涨跌幅度
stock=stock.iloc[::-1]
stock.head()
```

```python
stock=web.DataReader('^DJI','stooq',start,end)
stock.iloc[::-1]
```

```python
stock.describe()
```

```python
stock.index
```

```python
stock.columns
```

```python
stock.shape
```

```python
stock.sample(10)
```

```python
stock.info()
```

```python
change=stock.Close.diff()
stock['Change']=change
stock.head()
```

```python
stock.info()
```

```python
stock.describe()
```

```python
#确实数值用涨跌值得均值代替
stock['Change'].fillna(change.mean(),inplace=True)
stock.head()
```

```python
stock.describe()
```

```python
#绘制matplotlib

import matplotlib.pyplot as plt

#解决中文显示问题
plt.rcParams['font.sans-serif']="SimHei"
plt.rcParams["axes.unicode_minus"]=False
#设置图片标题
plt.title("{}至{}的DJI的Close数据图".format(start.strftime('%Y-%m-%d'),end.strftime('%Y-%m-%d')))
#对收盘价close列进行绘制，设置大小并有图例
stock['Close'].plot(legend=True,figsize=(10,4))
plt.show()
```

```python
#数据保存
stock.to_csv('DJI.csv',columns=stock.columns,index=True)
plt.savefig("{}至{}的DJI的Close数据图(部分).jpg".format(start.strftime('%Y-%m-%d'),end.strftime('%Y-%m-%d')))
```

# 1.3绘制图形

折线图： plot(x,y)

柱状图：bar（x,y）

饼图：pie(sizes,labels)

散点图：scatter（x,y）

直方图：hist（x,bins）

箱图：boxplot(x)

```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']="SimHei"
plt.rcParams["axes.unicode_minus"]=False

fig=plt.figure()
ax=fig.add_subplot(1,1,1)

sales = [100, 150, 120, 180, 210, 170]  
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']  
ax.bar(months,sales)
plt.show()
```

```python
ax.bar(sales,months)
plt.show()
```

```python
import matplotlib.pyplot as plt 
fig = plt.figure()  
ax = fig.add_subplot(1, 1, 1) 
sales = [100, 150, 120, 180, 210, 170]  
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']  
ax.plot(months,sales,linestyle='--',color='green',marker='o')
plt.show()
```
