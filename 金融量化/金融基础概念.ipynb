```python
def final_value(investment,rate,yeras):
    return investment*(1+rate)**yeras

investment=10000
rate=0.05
years=10
print('最终资产:',final_value(investment,rate,years))
```

# 影响收益率的核心要素

同等年限，收益率越高，最终资产越高

同等收益率，年限越长，最终资产越高

收益率是投资收益鱼投资成本之间的比值，反映投资收益与投资成本之间的比列

复利是指在计算利息时，某一计息周期的利息是由本金加上先前周期所积累利息总额来计算的计息方式

```python
#100元进行复利计算，年变化率保持在20%左右，50年之后变成多少呢？
print('复利20%，50年后资产为：',final_value(100,0.2,50))
print('存银行复利1.5%，50年后资产：',final_value(100,0.015,50))
```

```python
#可视化操作
inverstment=np.repeat(100,repeats=50)
rate=np.repeat(0.2,repeats=50)
years=np.arange(1,51)
rate_=np.repeat(0.015,repeats=50)

```

```python
date_20_50=pd.DataFrame(index=range(1,51),columns=['总资产'],data=inverstment*(1+rate)**years)
date_015_50=pd.DataFrame(index=range(1,51),columns=['存银行总资产'],data=inverstment*(1+rate_)**years)
plt.plot(date_20_50.index,date_20_50['总资产'],label='20%收益50年后总资产')
plt.plot(date_015_50.index,date_015_50['存银行总资产'],label='1.5%收益50年后总资产')
plt.legend()
plt.show()
```

# 最大回撤


```python
volatility_array=np.random.normal(size=100)/100
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# 计算最大回撤
def calculate_max_drawdown(prices):
    max_drawdown=0
    peak=prices[0]
    
    for price in prices:
        if price>peak:
            peak=price
        drawdown=(peak-price)/peak
        if drawdown>max_drawdown:
            max_drawdown=drawdown
    return max_drawdown
```

```python
# 计算最大回撤
#之前的收益率序列要改一下，转换为累计收益
volatility_array_cum=np.cumprod(1+volatility_array)
max_drawdown = calculate_max_drawdown(volatility_array_cum)
print("Max Drawdown: {:.2%}".format(max_drawdown))
#可视化累计收益走势，起始为1
plt.plot(np.arange(0,100),volatility_array_cum)
```
