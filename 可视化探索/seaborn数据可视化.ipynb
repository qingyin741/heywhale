```python
import seaborn as sns
import random 
import pandas as pd
import numpy as np
categories=['办公家具','厨房电器','家装饰品','床品件套','电脑硬件']
#使用列表生成式随机生成分类字段的值
random_categories=[random.choice(categories) for i in range(30)]
#导入数据
data=pd.DataFrame(data={
    '销售日期':pd.date_range('20231101',periods=30),
    '销售数':np.random.randint(low=100,high=1000,size=30),
    '销售单均':np.random.randint(low=50,high=100,size=30),
    '销售成本':np.random.randint(low=500,high=1000,size=30),
    '商品品类':random_categories
    })
data['日期']=data['销售日期'].dt.day
data['销售额']=data['销售数']*data['销售单均']
data['利润']=data['销售额']-data['销售成本']

new_columns=['销售日期', '日期', '销售数', '销售单均', '销售成本', '商品品类', '销售额', '利润']
data[new_columns]
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg' 
plt.figure(figsize=(9,6))
sns.lineplot(x='日期',y='销售数',data=data)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='svg'
plt.figure(figsize=(9,6))
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
sns.lineplot(x='日期',y='销售数',data=data,linestyle='--',linewidth=1,marker='o',markersize=5 )
```

```python
import seaborn as sns
#获取官方颜色转化成十六进制颜色值
colors=sns.color_palette().as_hex()

#依次打印出相关的值
for  idx,color in enumerate(colors):
    print(f'第{idx+1}种官方配色颜色值为:{color}')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='svg'
plt.figure(figsize=(9,6))
sns.lineplot(x='日期',y='销售数',data=data,color=sns.color_palette()[3])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='svg'
plt.figure(figsize=(9,6))
sns.lineplot(x='日期',y='销售数',data=data,hue='商品品类',
             hue_order=['电脑硬件','办公家具','家装饰品','床品件套','厨房电器'])
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.DataFrame({
    'date':pd.date_range('20231101',periods=30),
    'sale':np.random.randint(100,1000,size=30)
})

df['day']=df['date'].dt.day
plt.figure(figsize=(9,6))
sns.lineplot(x='day',y='sale',data=df,linestyle='--',linewidth=1,marker='o',markersize=5)
```

1. 柱状图展示不同类别之间的数值比较
2. 折线图适用于展示时间序列或连续变量之间的趋势关系
3. 散点图可以用来观察两个连续变量之间的关系
