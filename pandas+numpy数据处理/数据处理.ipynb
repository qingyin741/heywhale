```python


# Function to clean salary column
def salary_range_to_mean_fixed(salary):
    if isinstance(salary, str):
        salary = salary.replace('k', '')  # Remove 'k'
        if '-' in salary:
            low, high = salary.split('-')
            low, high = int(low) * 1000, int(high) * 1000  # Convert to actual numbers
            return (low + high) / 2
        return int(salary) * 1000
    return salary

# Apply the function to the salary column again
df['salary'] = df['salary'].apply(salary_range_to_mean_fixed)

# Now we calculate the mean of the min and max of the salary column
salary_max = df['salary'].max()
salary_min = df['salary'].min()
salary_mean = (salary_max + salary_min) / 2
df['salary'] = salary_mean

# Step 2: Calculate the range (max - min) and set it as 'ptp'
salary_range = salary_max - salary_min
df['ptp'] = salary_range

# Step 3: Categorize the salary into ['低', '中', '高']
bins = [0, 5000, 20000, 50000]
labels = ['低', '中', '高']
df['category'] = pd.cut(df['salary'], bins=bins, labels=labels)

# Step 4: Split createTime into 'date' and 'hour'
df['createTime'] = pd.to_datetime(df['createTime'])
df['date'] = df['createTime'].dt.strftime('%Y-%m-%d')
df['hour'] = df['createTime'].dt.hour

# Step 5: Filter data for 2020-03-16 and calculate the required statistics
filtered_df = df[df['date'] == '2020-03-16']

# Calculate required statistics
result = filtered_df.groupby(['date', 'hour']).agg(
    mean_salary=('salary', 'mean'),
    mean_ptp=('ptp', 'mean'),
    count_college=('education', lambda x: (x == '本科').sum()),
    count_master=('education', lambda x: (x == '硕士').sum()),
    count_low=('category', lambda x: (x == '低').sum()),
    count_meddle=('category', lambda x: (x == '中').sum()),
    count_high=('category', lambda x: (x == '高').sum())
).reset_index()

# Round the mean_salary to two decimal places
result['mean_salary'] = result['mean_salary'].round(2)

# Sort by date and hour
result = result.sort_values(by=['date', 'hour'])

# Display the updated DataFrame
result
```

```python
import pandas as pd

# 读取数据
file_path = r'E:\heywhale\零基础闯关\pandas120.xlsx'
df = pd.read_excel(file_path)

# 1. 将salary列数据转换为最大值与最小值的平均值
def convert_salary_to_avg(salary_str):
    try:
        # 分割薪资区间并去掉'k'
        salary_range = salary_str.replace('k', '').split('-')
        min_salary = int(salary_range[0]) * 1000
        max_salary = int(salary_range[1]) * 1000
        # 返回最大值与最小值的平均值
        return (min_salary + max_salary) / 2
    except:
        return None

# 应用函数处理salary列
df['salary_avg'] = df['salary'].apply(convert_salary_to_avg)

# 2. 计算salary列最大最小值之差（极差），设置列名为ptp
salary_max = df['salary_avg'].max()
salary_min = df['salary_avg'].min()
salary_ptp = salary_max - salary_min
df['ptp'] = salary_ptp

# 3. 新增一列根据salary列数值大小划分为三个层次['低', '中', '高']，列名命名category
bins = [0, 5000, 20000, 50000]
group_names = ['低', '中', '高']
df['category'] = pd.cut(df['salary_avg'], bins, labels=group_names)

# 4. 根据createTime列，拆分两列字段:日期(年-月-日格式)、小时，分别命名date，hour
df['createTime'] = pd.to_datetime(df['createTime'])
df['date'] = df['createTime'].dt.date
df['hour'] = df['createTime'].dt.hour

# 5. 统计在2020-03-16这一天，每个小时的平均工资、平均极差(每行数据极差都是41500)、本科和硕士学历个数，薪资层次中高中低的个数
# 过滤出2020-03-16的数据
df_filtered = df[df['date'] == pd.to_datetime('2020-03-16').date()]

# 按照小时分组，计算所需统计数据
result = df_filtered.groupby('hour').agg(
    mean_salary=('salary_avg', lambda x: round(x.mean(), 2)),  # 平均工资保留两位小数
    mean_ptp=('ptp', 'mean'),  # 平均极差
    count_college=('education', lambda x: (x == '本科').sum()),  # 本科学历个数
    count_master=('education', lambda x: (x == '硕士').sum()),  # 硕士学历个数
    count_low=('category', lambda x: (x == '低').sum()),  # 低薪资层次个数
    count_meddle=('category', lambda x: (x == '中').sum()),  # 中薪资层次个数
    count_high=('category', lambda x: (x == '高').sum())  # 高薪资层次个数
).reset_index()

# 添加日期列
result['date'] = pd.to_datetime('2020-03-16').date()

# 按照date和hour升序排序
result = result[['date', 'hour', 'mean_salary', 'mean_ptp', 'count_college', 'count_master', 'count_low', 'count_meddle', 'count_high']].sort_values(by=['date', 'hour'])

# 输出最终结果
result
```
