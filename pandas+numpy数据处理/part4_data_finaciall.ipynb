```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the Excel file
file_path = r'E:\heywhale\pandas+numpy数据处理\600000.SH.xlsx'
data = pd.read_excel(file_path)

# Step 2: Drop rows with any missing values
data.dropna(axis=0, how='any', inplace=True)

# Step 3: Remove rows where '换手率(%)' is not a numeric value
non_numeric_rows = data[~data['换手率(%)'].apply(lambda x: isinstance(x, (int, float)))].index
data.drop(non_numeric_rows, inplace=True)

# Step 4: Set the '日期' (Date) as the index
data['日期'] = pd.to_datetime(data['日期'])
data.set_index('日期', inplace=True)

# Step 5: Shift the data forward and backward by 5 days
data_shifted_forward = data.shift(-5)
data_shifted_backward = data.shift(5)

# Step 6: Calculate expanding window mean of the opening price ('开盘价(元)')
data['Expanding Mean 开盘价(元)'] = data['开盘价(元)'].expanding().mean()

# Step 7: Plot the expanding mean along with the original opening price
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['开盘价(元)'], label='Original 开盘价(元)', color='blue')
plt.plot(data.index, data['Expanding Mean 开盘价(元)'], label='Expanding Mean 开盘价(元)', color='red')
plt.title('Opening Price and Expanding Mean')
plt.xlabel('Date')
plt.ylabel('Price (元)')
plt.grid(True)
plt.legend()
plt.show()

# Step 8: Calculate Bollinger Bands
N = 20  # Define the window for rolling calculations
data['MA'] = data['收盘价(元)'].rolling(window=N).mean()  # Moving average
data['MD'] = data['收盘价(元)'].rolling(window=N).std()   # Standard deviation
data['UP'] = data['MA'] + (2 * data['MD'])  # Upper band
data['DN'] = data['MA'] - (2 * data['MD'])  # Lower band

# Step 9: Plot the closing price with Bollinger Bands
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['收盘价(元)'], label='收盘价(元)', color='blue')
plt.plot(data.index, data['MA'], label='Middle Band (MA)', color='red')
plt.plot(data.index, data['UP'], label='Upper Band (UP)', color='green')
plt.plot(data.index, data['DN'], label='Lower Band (DN)', color='orange')
plt.title('Bollinger Bands and Closing Price')
plt.xlabel('Date')
plt.ylabel('Price (元)')
plt.grid(True)
plt.legend()
plt.show()

```

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

data=pd.read_excel(r'E:\heywhale\pandas+numpy数据处理\600000.SH.xls')
# 删除空值
data.dropna(axis=0,how='any',inplace=True)
# 删除非数值行
temp=[]
for i in range(len(data)):
    if type(data['换手率(%)'][i]) !=float:
        temp.append(i)
data.drop(labels=temp,inplace=True)
data=data.set_index('日期')
# 将数据往前移动5填
data=data.shift(-5)

#使用expanding函数计算开盘价
raw_data=data['开盘价(元)']
expanding_data=data['开盘价(元)'].expanding(min_periods=1,axis=0).mean()

#将两个数据拼接到一起
result=pd.concat([raw_data,expanding_data],axis=1,ignore_index=True)
result= result.rename(columns={0: '原始数据',   1: '移动窗口均值'})


#绘制图
plt.figure('figure',figsize=(16,6),dpi=60)
plt.plot(result['原始数据'])
plt.plot(result['移动窗口均值'])

#布林线指标参数设置为20
N = 20
# MA移动平均线
MA=data['收盘价(元)'].rolling(N).mean()

# Md  计算标准差
MD=data['收盘价(元)'].rolling(N).std()

MB=MA
UP=MB+(MD*2)
DN=MB-(MD*2)

# 计算布林线并绘制
plt.rc('figure',figsize=(16,6),dpi=50)
plt.plot(MA,color='blue')
plt.plot(UP,color='red')
plt.plot(DN,color='g')
plt.show()
dic = {'上轨线': UP}
df = pd.DataFrame(dic).reset_index(drop=True)
df = df.round(2)
df.columns = ['answer']
df.dropna(axis=0, how='any', inplace=True)
df['id'] = range(len(df))
df = df[['id', 'answer']]
df
```
