```python
import pandas as pd

# Creating the initial DataFrame
data = {
    "Id": [202001, 202002, 202003, 202004, 202005, 202006, 202007, 202008, 202009, 202010],
    "Chinese": [98, 67, 84, 88, 78, 90, 93, 75, 82, 87],
    "Math": [92, 80, 73, 76, 88, 78, 90, 82, 77, 69],
    "English": [88, 79, 90, 73, 79, 83, 81, 91, 71, 78]
}

df = pd.DataFrame(data)

# Calculating the required metrics
df["SumScore"] = df[["Chinese", "Math", "English"]].sum(axis=1)
df["MeanScore"] = df[["Chinese", "Math", "English"]].mean(axis=1).astype(int)
df["MaxScore"] = df[["Chinese", "Math", "English"]].max(axis=1)
df["MinScore"] = df[["Chinese", "Math", "English"]].min(axis=1)
df["PtpScore"] = df["MaxScore"] - df["MinScore"]
df["VarScore"] = df[["Chinese", "Math", "English"]].var(axis=1,ddof=0).astype(int)

StudentScore=df.copy()
data = pd.concat([StudentScore.iloc[:,0],StudentScore.iloc[:,1],StudentScore.iloc[:,2],StudentScore.iloc[:,3],StudentScore.iloc[:,4],StudentScore.iloc[:,5],StudentScore.iloc[:,6],StudentScore.iloc[:,7],StudentScore.iloc[:,8],StudentScore.iloc[:,9]])
df = pd.DataFrame(data, columns=['answer'])
df['id'] = range(len(df))
df = df[['id', 'answer']]
df.tail(20)
```
