```python
import sqlite3
from datetime import datetime, timedelta

# åˆ›å»º SQLite æ•°æ®åº“è¿æ¥
conn = sqlite3.connect(":memory:")
cursor = conn.cursor()

# åˆ›å»ºæ—¥å†è¡¨
cursor.execute("""
CREATE TABLE calendar_table (
    date TEXT,
    year INTEGER,
    month INTEGER,
    day INTEGER,
    week INTEGER
)
""")

# å¡«å……æ•°æ®ï¼ˆ2023å¹´ï¼‰
start_date = datetime(2023, 1, 1)
end_date = datetime(2023, 12, 31)
current_date = start_date

while current_date <= end_date:
    year = current_date.year
    month = current_date.month
    day = current_date.day
    
    # ISO å‘¨è®¡ç®—ï¼Œæ¯å‘¨ä»å‘¨ä¸€å¼€å§‹
    week = (day + (current_date.weekday() - 1) // 7) if day else 1
    cursor.execute("""
    INSERT INTO calendar_table (date, year, month, day, week)
    VALUES (?, ?, ?, ?, ?)
    """, (current_date.strftime("%Y-%m-%d"), year, month, day, week))
    current_date += timedelta(days=1)

conn.commit()

# æŸ¥è¯¢ç›®æ ‡æ•°æ®
query = """
WITH fifth_day_weeks AS (
    SELECT 
        month,
        week
    FROM 
        calendar_table
    WHERE 
        day = 5
),
longest_streak AS (
    SELECT 
        week,
        COUNT(*) AS consecutive_count
    FROM (
        SELECT 
            month,
            week,
            week = LAG(week) OVER (ORDER BY month) AS is_consecutive
        FROM 
            fifth_day_weeks
    ) consecutive_weeks
    WHERE is_consecutive = true
    GROUP BY week
)
SELECT 
    MAX(consecutive_count) AS max_streak
FROM 
    longest_streak;
"""

# æ‰§è¡ŒæŸ¥è¯¢
cursor.execute(query)
result = cursor.fetchone()

# è¾“å‡ºç»“æœ
print(f"2023 å¹´æœ€é•¿ä¸€æ¬¡è¿ç»­çš„ 5 å·å¤„äºåŒä¸€å‘¨çš„æœˆä»½æ•°ä¸ºï¼š{result[0]}")

# å…³é—­æ•°æ®åº“è¿æ¥
conn.close()
```

```python
import pandas as pd
from datetime import datetime, timedelta

# æ„é€ æ•°æ®
events = [
    {"event": "æ‰“ç£¨æ–°çš„å…µå™¨", "start_date": "2023-02-10", "duration": 17, "importance": 2},
    {"event": "çœ‹1æœˆæ–°ç•ª", "start_date": "2023-05-02", "duration": 8, "importance": 3},
    {"event": "ä¿®è¡¥è¢«ç ´åçš„åŸå ¡", "start_date": "2023-02-23", "duration": 31, "importance": 1},
    {"event": "å®‰æ’å°æ¶é­”å»æ”¶ä¿æŠ¤è´¹", "start_date": "2023-03-17", "duration": 14, "importance": 0},
    # æ·»åŠ å…¶ä½™äº‹ä»¶æ•°æ®...
]
events_df = pd.DataFrame(events)

# ç”Ÿæˆ 2023 å¹´æ¯æ—¥æ•°æ®
dates = pd.date_range(start="2023-01-01", end="2023-12-31")
calendar_df = pd.DataFrame({"date": dates})

# è®¡ç®—ç¹å¿™ç³»æ•°
def compute_busy_factor(date, events_df):
    # å½“å¤©çš„äº‹é¡¹
    current_day_events = events_df[
        (events_df["start_date"] <= date) &
        (events_df["start_date"] + pd.to_timedelta(events_df["duration"], unit="d") >= date)
    ]
    avg_importance_today = current_day_events["importance"].mean() if not current_day_events.empty else 0

    # æœªæ¥ 3 å¤©å†…ç»“æŸçš„äº‹é¡¹
    ending_3days_events = events_df[
        (events_df["start_date"] + pd.to_timedelta(events_df["duration"], unit="d") > date) &
        (events_df["start_date"] + pd.to_timedelta(events_df["duration"], unit="d") <= date + timedelta(days=3))
    ]
    avg_importance_ending_3days = ending_3days_events["importance"].mean() if not ending_3days_events.empty else 0

    # æœªæ¥ 3 å¤©å†…å¼€å§‹çš„äº‹é¡¹
    starting_3days_events = events_df[
        (events_df["start_date"] > date) &
        (events_df["start_date"] <= date + timedelta(days=3))
    ]
    avg_importance_starting_3days = starting_3days_events["importance"].mean() if not starting_3days_events.empty else 0

    return avg_importance_today * 0.5 + avg_importance_ending_3days * 0.2 + avg_importance_starting_3days * 0.3

# æ·»åŠ ç¹å¿™ç³»æ•°åˆ°æ¯æ—¥æ•°æ®
calendar_df["busy_factor"] = calendar_df["date"].apply(lambda d: compute_busy_factor(d, events_df))

# ç­›é€‰ç¹å¿™çš„å¤©æ•°
busy_days = calendar_df[
    (calendar_df["date"].apply(
        lambda d: len(events_df[
            (events_df["start_date"] <= d) &
            (events_df["start_date"] + pd.to_timedelta(events_df["duration"], unit="d") >= d)
        ]) >= 2
    )) |
    (calendar_df["date"].apply(
        lambda d: len(events_df[
            (events_df["start_date"] > d) & (events_df["start_date"] <= d + timedelta(days=3)) & (events_df["importance"] > 2)
        ]) > 0
    )) |
    (calendar_df["date"].apply(
        lambda d: len(events_df[
            (events_df["start_date"] + pd.to_timedelta(events_df["duration"], unit="d") > d) &
            (events_df["start_date"] + pd.to_timedelta(events_df["duration"], unit="d") <= d + timedelta(days=3)) & (events_df["importance"] > 1)
        ]) > 0
    ))
]

# è®¡ç®—æ€»ç¹å¿™ç³»æ•°
busy_factor = busy_days["busy_factor"].sum()
print(f"ç¹å¿™ç³»æ•°ï¼ˆæ•´æ•°ä½ï¼‰: {int(busy_factor)}")
```
