# Q1:2023年星期二和星期五的关系 天数是否相等

```python
# 检查星期二和星期五的关系
import calendar
year=2023

#统计每个星期出现的次数
week_counts=[0]*7

# 遍历每个月的每一天
for month in range(1,13):
    month_calendar=calendar.monthcalendar(year,month) #获取月份周期
    print(month_calendar)
    for week in month_calendar:
        print("week is ",week)
        for day,weekday in enumerate(week):   #可以理解为创建了索引对应
            print(day,weekday)
            if weekday !=0:         # 结束
                week_counts[day]+=1
                
weekdays=['星期一','星期二','星期三','星期四','星期五','星期六','星期日']
for i ,count in enumerate(week_counts):
    print(f"{weekdays[i]}:{count}次")
# 检查星期二和星期五的关系
if week_counts[1]>week_counts[4]:
    print(">")
elif week_counts[1]==week_counts[4]:
    print('=')
else:
    print('<')
```

# Q2: 在 2023 年每个月的5号，10号，15号，20号和 25号 中，星期几的占比最高¶

```python
import calendar
from collections import Counter
year=2023
#目标日期
target_days=[5,10,15,20,25]
#存储每个目标日期对应的星期
weekday_counts=[]
#遍历每个月
for month in range(1,13):
    for day in target_days:
        try:
            # 获取某一天的星期几
            weekday=calendar.weekday(year,month,day)
            weekday_counts.append(weekday)
        except ValueError:
            continue

#统计星期几出现的次数
weekday_counter=Counter(weekday_counts)
print(weekday_counter)
weekdays=['星期一', '星期二', '星期三', '星期四', '星期五', '星期六', '星期日']       
result={weekdays[day]:count for day ,count in weekday_counter.items()}
max_day=max(result,key=result.get)
print('2023年目标日期中各星期的统计：',result)
print(f"占比最高的星期是{max_day},出现次数是{result[max_day]}")
```

# Q4: 在 2023 年所有月份的最后 1 周，一共有多少个周末  


```python
import  calendar
year=2023
#周末统计
total_weekends=0
#遍历每个月
for month in range(1,13):
    #获取该月最后一天
    _,last_day=calendar.monthrange(year,month)
    #获取最后一周的起始日期
    last_week_start=last_day-6
    
    for day in range(last_week_start,last_day+1):
        weekday=calendar.weekday(year,month,day)
        if weekday in [5,6]:
            total_weekends+=1
        
print(f"2023年所有月份的最后一周共有{total_weekends}个周末")
        
    
```

# Q5: 在 2023 年所有月份的 5 号，最长一次连续处在当月同一周是连续几个月？
例如：2,3,4,5月份的 5 号都处于当月的第 2 周， 则最长一次为 4

```python
import calendar

year=2023
# 存储每个月的5号所在周期
weeks=[]
for month  in range(1,13):
    week_num=(5+calendar.monthrange(year,month)[0]-1)//7+1
    weeks.append(week_num)

#找最长连续相同周数的月份
max_streak=1
current_streak=1

for i in range(1,len(weeks)):
    if weeks[i]==weeks[i-1]:
        current_streak+=1
        max_streak=max(max_streak,current_streak)
    else:
        current_streak=1
        
print(f"2023年最长一次连续的5号处于同一周的月份数为:{max_streak}")
```

```python
import sqlite3
from datetime import datetime, timedelta

# 创建 SQLite 数据库连接
conn = sqlite3.connect(":memory:")
cursor = conn.cursor()

# 创建日历表
cursor.execute("""
CREATE TABLE calendar_table (
    date TEXT,
    year INTEGER,
    month INTEGER,
    day INTEGER,
    week INTEGER
)
""")

# 填充数据（2023年）
start_date = datetime(2023, 1, 1)
end_date = datetime(2023, 12, 31)
current_date = start_date

while current_date <= end_date:
    year = current_date.year
    month = current_date.month
    day = current_date.day
    
    # ISO 周计算，每周从周一开始
    week = (day + (current_date.weekday() - 1) // 7) if day else 1
    cursor.execute("""
    INSERT INTO calendar_table (date, year, month, day, week)
    VALUES (?, ?, ?, ?, ?)
    """, (current_date.strftime("%Y-%m-%d"), year, month, day, week))
    current_date += timedelta(days=1)

conn.commit()

# 查询目标数据
query = """
WITH fifth_day_weeks AS (
    SELECT 
        month,
        week
    FROM 
        calendar_table
    WHERE 
        day = 5
),
longest_streak AS (
    SELECT 
        week,
        COUNT(*) AS consecutive_count
    FROM (
        SELECT 
            month,
            week,
            week = LAG(week) OVER (ORDER BY month) AS is_consecutive
        FROM 
            fifth_day_weeks
    ) consecutive_weeks
    WHERE is_consecutive = true
    GROUP BY week
)
SELECT 
    MAX(consecutive_count) AS max_streak
FROM 
    longest_streak;
"""

# 执行查询
cursor.execute(query)
result = cursor.fetchone()

# 输出结果
print(f"2023 年最长一次连续的 5 号处于同一周的月份数为：{result[0]}")

# 关闭数据库连接
conn.close()
```

```python
import pandas as pd
from datetime import datetime, timedelta

# 构造数据
events = [
    {"event": "打磨新的兵器", "start_date": "2023-02-10", "duration": 17, "importance": 2},
    {"event": "看1月新番", "start_date": "2023-05-02", "duration": 8, "importance": 3},
    {"event": "修补被破坏的城堡", "start_date": "2023-02-23", "duration": 31, "importance": 1},
    {"event": "安排小恶魔去收保护费", "start_date": "2023-03-17", "duration": 14, "importance": 0},
    # 添加其余事件数据...
]
events_df = pd.DataFrame(events)

# 生成 2023 年每日数据
dates = pd.date_range(start="2023-01-01", end="2023-12-31")
calendar_df = pd.DataFrame({"date": dates})

# 计算繁忙系数
def compute_busy_factor(date, events_df):
    # 当天的事项
    current_day_events = events_df[
        (events_df["start_date"] <= date) &
        (events_df["start_date"] + pd.to_timedelta(events_df["duration"], unit="d") >= date)
    ]
    avg_importance_today = current_day_events["importance"].mean() if not current_day_events.empty else 0

    # 未来 3 天内结束的事项
    ending_3days_events = events_df[
        (events_df["start_date"] + pd.to_timedelta(events_df["duration"], unit="d") > date) &
        (events_df["start_date"] + pd.to_timedelta(events_df["duration"], unit="d") <= date + timedelta(days=3))
    ]
    avg_importance_ending_3days = ending_3days_events["importance"].mean() if not ending_3days_events.empty else 0

    # 未来 3 天内开始的事项
    starting_3days_events = events_df[
        (events_df["start_date"] > date) &
        (events_df["start_date"] <= date + timedelta(days=3))
    ]
    avg_importance_starting_3days = starting_3days_events["importance"].mean() if not starting_3days_events.empty else 0

    return avg_importance_today * 0.5 + avg_importance_ending_3days * 0.2 + avg_importance_starting_3days * 0.3

# 添加繁忙系数到每日数据
calendar_df["busy_factor"] = calendar_df["date"].apply(lambda d: compute_busy_factor(d, events_df))

# 筛选繁忙的天数
busy_days = calendar_df[
    (calendar_df["date"].apply(
        lambda d: len(events_df[
            (events_df["start_date"] <= d) &
            (events_df["start_date"] + pd.to_timedelta(events_df["duration"], unit="d") >= d)
        ]) >= 2
    )) |
    (calendar_df["date"].apply(
        lambda d: len(events_df[
            (events_df["start_date"] > d) & (events_df["start_date"] <= d + timedelta(days=3)) & (events_df["importance"] > 2)
        ]) > 0
    )) |
    (calendar_df["date"].apply(
        lambda d: len(events_df[
            (events_df["start_date"] + pd.to_timedelta(events_df["duration"], unit="d") > d) &
            (events_df["start_date"] + pd.to_timedelta(events_df["duration"], unit="d") <= d + timedelta(days=3)) & (events_df["importance"] > 1)
        ]) > 0
    ))
]

# 计算总繁忙系数
busy_factor = busy_days["busy_factor"].sum()
print(f"繁忙系数（整数位）: {int(busy_factor)}")
```
